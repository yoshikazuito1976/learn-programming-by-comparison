# Python input practice

このノートブックでは、Pythonで外部からデータを受け取る方法を確認します。

扱う内容は次の通りです。

- 固定値を使うプログラム
- `input()` による標準入力
- 入力された値の型
- 数値への変換
- 繰り返し入力
- ファイルからの入力
- `with open()` の基本

まずは、プログラムの外から値を受け取る感覚をつかみましょう。

## 1. 固定値を使うプログラム

まずは、プログラムの中にあらかじめ値を書いておく例です。

```python
name = "Yoshikazu"

print(f"こんにちは、{name}さん")
```

この場合、`name` の値はプログラムの中で決まっています。

毎回違う名前を使いたい場合は、プログラムを書き換える必要があります。

## 2. input() でキーボードから入力する

`input()` を使うと、キーボードから入力された文字を受け取ることができます。

次のセルを実行すると、入力欄が表示されます。
名前を入力して Enter を押してください。

```python
name = input("名前を入力してください: ")

print(f"こんにちは、{name}さん")
```

## 3. input() で受け取った値の型

`input()` で受け取った値は、基本的に文字列です。

`type()` を使って確認してみましょう。

```python
age = input("年齢を入力してください: ")

print(age)
print(type(age))
```

たとえば `20` と入力しても、`input()` で受け取った値は数値ではなく文字列です。

そのため、このままでは数値計算に使えません。

## 4. あえてエラーを確認する

次のセルは、あえてエラーになる例です。

`input()` で受け取った文字列に、数値の `1` を足そうとしているためです。

```python
age = input("年齢を入力してください: ")

print(age + 1)
```


エラーが出ても問題ありません。

ここでは、`input()` の戻り値が文字列であることを確認するために、あえてエラーを出しています。

## 5. int() で数値に変換する

入力された文字列を整数として扱いたい場合は、`int()` を使います。

```python
age_text = input("年齢を入力してください: ")
age = int(age_text)

print(f"現在は {age} 歳です")
print(f"来年は {age + 1} 歳です")
```

`input()` で受け取る

↓

`int()` で数値に変換する

↓

計算に使う

という流れを意識しましょう。

## 6. for文とinput()を組み合わせる

繰り返し処理と `input()` を組み合わせると、複数の値を受け取ることができます。

ここでは、3人分の点数を入力して、合計点と平均点を計算します。

```python
scores = []

for i in range(3):
    score = int(input(f"{i + 1}人目の点数を入力してください: "))
    scores.append(score)

print(scores)
print(f"合計点: {sum(scores)}")
print(f"平均点: {sum(scores) / len(scores)}")
```

この例では、次の処理をしています。

1. 空のリスト `scores` を用意する
2. `for` 文で3回繰り返す
3. 入力された点数を `int()` で数値に変換する
4. `append()` でリストに追加する
5. `sum()` と `len()` で合計点と平均点を計算する

## 7. ファイルを作成する

次に、ファイルから入力する例を確認します。

まず、練習用の `sample.txt` を作成します。

```python
with open("sample.txt", "w", encoding="utf-8") as file:
    file.write("apple\n")
    file.write("banana\n")
    file.write("orange\n")
```

`"w"` は書き込みモードです。

同じ名前のファイルがすでにある場合、中身は上書きされます。

## 8. ファイル全体を読み込む

作成した `sample.txt` を読み込んでみましょう。

```python
with open("sample.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text)
```

`"r"` は読み込みモードです。

`read()` を使うと、ファイル全体を1つの文字列として読み込めます。

## 9. ファイルを1行ずつ読み込む

ファイルは、1行ずつ読み込むこともできます。

```python
with open("sample.txt", "r", encoding="utf-8") as file:
    for line in file:
        word = line.strip()
        print(f"読み込んだ単語: {word}")
```

`strip()` は、文字列の前後にある空白や改行を取り除くメソッドです。

ファイルの1行には、末尾に改行が含まれていることがあります。
そのため、1行ずつ処理するときには `strip()` を使うことがあります。

## 10. open() のモード

`open()` の第2引数には、ファイルをどのように開くかを指定します。

| モード | 意味 | ファイルがない場合 | 既存内容 |
|---|---|---|---|
| `"r"` | 読み込み | エラー | 残る |
| `"w"` | 書き込み | 新規作成 | 消える |
| `"a"` | 追記 | 新規作成 | 残る |
| `"x"` | 新規作成して書き込み | 新規作成 | 既にあるとエラー |

## 11. 追記モードを使う

`"a"` は追記モードです。

既存の内容を消さずに、ファイルの末尾に追加します。

```python
with open("sample.txt", "a", encoding="utf-8") as file:
    file.write("grape\n")

with open("sample.txt", "r", encoding="utf-8") as file:
    text = file.read()

print(text)
```

## 12. ファイルを開く・閉じる

ファイル操作は、基本的に次の流れで考えます。

1. ファイルを開く
2. ファイルを読む、または書く
3. ファイルを閉じる

`with open(...) as file:` を使うと、ファイルを使い終わったあとにPythonが自動的に閉じてくれます。

## 13. close() を自分で書く場合

次のように、`close()` を自分で書くこともできます。

```python
file = open("sample.txt", "r", encoding="utf-8")
text = file.read()
file.close()

print(text)
```

ただし、この書き方では `close()` を書き忘れる可能性があります。

そのため、Pythonでは基本的に `with open(...) as file:` を使うのが安全です。
